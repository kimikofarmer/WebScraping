```python
#import packages
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import re
import itertools
```

```python
url="http://krieger.jhu.edu/east-asian/people/" #url of interest
req = urllib.request.Request(url, headers={'User-Agent' : "Magic Browser"}) 
page = urllib.request.urlopen( req )
soup = BeautifulSoup(page.read(), 'lxml') #parse page using xml parser
```

```python
#initialize lists
names = []
phdUniversity = []
phdUniversityCode = []
currentUniversity = []
currentUniversityCode = []
currentDepartment = []
```

```python
#get list of all the faculty
people = soup.find_all("li", { "class" : "person faculty" }) + soup.find_all("li", { "class" : "person adjunct-and-affiliated-faculty" })  

#for each faculty member, find their name and biography
for each in people:
    bio = each.find('h5', attrs={'class': 'subheader'})
    name = each.find('h4', attrs={'class': 'no-margin'})
    name = [x.strip() for x in name.get_text().split('\n') if x.strip()]
    
    #if a unique person has a bio continue
    if bio is not None and name[0] not in names:
        text = [x.strip() for x in bio.get_text().split('\t') if x.strip()]
        school = text[-1]
        school = school.split(", ")
        
        #check to make sure they have their PhD, then append info
        if(school[0] == 'PhD'):
            names.append(name[0])
            phdUniversity.append(school[1])
            currentDepartment.append("East Asian Studies")
```

```python
url="http://engineering.jhu.edu/chembe/faculty/" #url of interest
req = urllib.request.Request(url, headers={'User-Agent' : "Magic Browser"}) 
page = urllib.request.urlopen( req )
soup = BeautifulSoup(page.read(), 'lxml') #parse page using xml parser
```

```python
#get list of all faculty
people = soup.find_all("div", { "class" : "highlight-box faculty-member-card" })

#for each faculty member, find their name and biography
for each in people:
    bio = each.find('p')
    name = each.find('h3')
    name = [x.strip() for x in name.get_text().split('\n') if x.strip()]
    
    #if a unique person has a bio continue
    if bio is not None and name[0] not in names:    
        text = [x.strip() for x in bio.get_text().split('\n') if x.strip()]
        school = text[-1]
        school = school.split(", ")
        
        #check to make sure they have their PhD, then append info
        if re.match('^Ph.D.', school[0]) is not None:
            university = school[-1]
            university = university[:-6]
            if re.match('^Ph.D.', university) is not None:
                university = university[6:]
            names.append(name[0])
            phdUniversity.append(university)
            currentDepartment.append("Chemical and Biomolecular Engineering")
```

```python
#split names into first and last name
#last name is (assumed to be) last element of split at white space
lastNames = [x.split()[-1] for x in names]
firstNames = [x.split()[0] for x in names]
```

```python
#imports the school codes csv and makes a dictionary
schoolCodes = pd.read_csv("school_codes.csv")
schoolID = schoolCodes["co"].tolist()
schoolName = schoolCodes["name"].tolist()
schoolDictionary = dict(zip(schoolName,schoolID))

#for each phd university, find and append its school code
for each in phdUniversity:
    if each in schoolDictionary:
        phdUniversityCode.append(schoolDictionary.get(each))
    
    #special cases: they are in the school codes csv, but as a different name
    else:
        if(each=="UCLA"):
            phdUniversityCode.append(schoolDictionary.get("UC Los Angeles"))
        elif(each=="SUNY-Binghamton"):
            phdUniversityCode.append(schoolDictionary.get("SUNY Binghamton"))
        elif(each=="University of California"):
            phdUniversityCode.append(schoolDictionary.get("UC Berkeley"))
        elif(each=="University of California at Berkeley"):
            phdUniversityCode.append(schoolDictionary.get("UC Berkeley"))
        elif(each=="University of Maryland"):
            phdUniversityCode.append(schoolDictionary.get("University of Maryland, Baltimore"))
        elif(each=="Rutgers University-New Brunswick"):
            phdUniversityCode.append(schoolDictionary.get("Rutgers University, New Brunswick"))
        elif(each=="Indiana University Bloomington"):
            phdUniversityCode.append(schoolDictionary.get("Indiana University, Bloomington"))
        elif(each=="Oxford University"):
            phdUniversityCode.append(schoolDictionary.get("University of Oxford"))
        elif(each=="Ohio State University"):
            phdUniversityCode.append(schoolDictionary.get("Ohio State University, Columbus"))
        elif(each=="University of Michigan"):
            phdUniversityCode.append(schoolDictionary.get("University of Michigan, Ann Arbor"))
        elif(each=="University of Texas at Austin"):
            phdUniversityCode.append(schoolDictionary.get("University of Texas, Austin"))
        elif(each=="San Francisco"):
            phdUniversityCode.append(schoolDictionary.get("UC San Francisco"))
```

```python
#retrieved the number of faculty and filled in blank columns and faculty university id 
#Johns Hopkins ID = 244
length = len(phdUniversity)
currentUniversity = (["Johns Hopkins University"] * length)
currentUniversityCode = (["244"] * length)
phdDepartment = ([""] * length)
phdYear = ([""] * length)
yearStarted = ([""] * length)
position = ([""] * length)

#formatting the data frame
data = pd.DataFrame({'First Name': firstNames, 'Last name': lastNames, 'university of PhD': phdUniversity, 'university of PhD ID': phdUniversityCode, 'department of PhD': phdDepartment, 'year of PhD': phdYear, 'current university': currentUniversity, 'current university ID': currentUniversityCode, 'department of current faculty': currentDepartment, 'year faculty started': yearStarted, 'job title': position}) #format into my DF
data = data[['First Name','Last name', 'university of PhD', 'university of PhD ID', 'department of PhD', 'year of PhD', 'current university', 'current university ID', 'department of current faculty', 'year faculty started', 'job title']]
```
